```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

# ─────────────────────────────────────────────────────────────────────────────
# 1. LOAD DATA
# ─────────────────────────────────────────────────────────────────────────────
df_bus        = pd.read_csv('clean_bus.csv')
df_sub        = pd.read_csv('clean_sub.csv')
df_streetcar  = pd.read_csv('clean_streetcar.csv')

# ─────────────────────────────────────────────────────────────────────────────
# 2. FEATURE ENGINEERING
#    applied to all three datasets
# ─────────────────────────────────────────────────────────────────────────────
def engineer_features(df):
    df = df.copy()

    # normalize column name — some files use Line, others Route
    if 'Line' in df.columns and 'Route' not in df.columns:
        df = df.rename(columns={'Line': 'Route'})
    
    # ... rest of function unchanged

    # time flags
    df['IsRushHour'] = np.where(
        df['Hour'].between(7, 9) | df['Hour'].between(16, 18), 1, 0
    )
    df['IsWeekend'] = (df['DayOfWeek'] >= 5).astype(int)
    df['IsNight']   = np.where(df['Hour'].between(0, 5), 1, 0)

    # season
    conditions = [
        df['Month'].isin([12, 1, 2]),
        df['Month'].between(3, 5),
        df['Month'].between(6, 8),
        df['Month'].between(9, 11),
    ]
    df['Season'] = np.select(
    conditions,
    ['Winter', 'Spring', 'Summer', 'Fall'],
    default='Unknown'
)

    # drop COVID years
    df = df[df['Year'] >= 2023].copy()

    # route-level aggregates
    route_stats = df.groupby('Route')['Min Delay'].agg(
        RouteAvgDelay='mean',
        RouteDelayStd='std'
    ).reset_index()
    df = df.merge(route_stats, on='Route', how='left')

    # gap deviation from route average
    route_avg_gap       = df.groupby('Route')['Min Gap'].transform('mean')
    df['GapDeviation']  = df['Min Gap'] - route_avg_gap

    # vehicle previous delay (sort by vehicle + time first)
    df = df.sort_values(['Vehicle', 'Year', 'Month', 'DayOfWeek', 'Hour'])
    df['VehiclePrevDelay'] = df.groupby('Vehicle')['Min Delay'].shift(1)
    df['VehiclePrevDelay'] = df['VehiclePrevDelay'].fillna(df['RouteAvgDelay'])

    return df


df_bus       = engineer_features(df_bus)
df_sub       = engineer_features(df_sub)
df_streetcar = engineer_features(df_streetcar)

# ─────────────────────────────────────────────────────────────────────────────
# 3. CONFIG
# ─────────────────────────────────────────────────────────────────────────────
datasets = {
    'Bus':       df_bus,
    'Subway':    df_sub,
    'Streetcar': df_streetcar,
}

TARGET = 'Min Delay'

FEATURES = [
    'Hour', 'Month', 'DayOfWeek',
    'IsRushHour', 'IsWeekend', 'IsNight',
    'Direction_known',
    'Season', 'Incident',
    'Vehicle', 'Route',
    'RouteAvgDelay', 'RouteDelayStd',
    'GapDeviation', 'VehiclePrevDelay',
    'Min Gap',
]
FEATURES_NO_GAP = [f for f in FEATURES if f not in ['Min Gap', 'GapDeviation']]

# cost per delay minute by mode — from TTC annual report
COST_PER_MIN = {
    'Bus':       1.23,
    'Subway':    2.87,
    'Streetcar': 1.95,
}

# ─────────────────────────────────────────────────────────────────────────────
# 4. ENCODE CATEGORICALS
# ─────────────────────────────────────────────────────────────────────────────
def encode(df, encoders=None, fit=True):
    df = df.copy()
    if encoders is None:
        encoders = {}
    for col in df.select_dtypes(include='object').columns:
        if col == TARGET:
            continue
        if fit:
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col].astype(str))
            encoders[col] = le
        else:
            le = encoders[col]
            # handle unseen labels gracefully
            df[col] = df[col].astype(str).map(
                lambda x: le.transform([x])[0] if x in le.classes_ else -1
            )
    return df, encoders

# ─────────────────────────────────────────────────────────────────────────────
# 5. TRAIN + EVALUATE
# ─────────────────────────────────────────────────────────────────────────────
results = {}

for name, df in datasets.items():
    print(f"\n{'='*54}")
    print(f"  {name}")
    print(f"{'='*54}")

    # keep only columns that exist in this dataframe
    features = [f for f in FEATURES_NO_GAP if f in df.columns]
    
    data = df[features + [TARGET, 'Year']].dropna()

    # time-based split — no data leakage
    train = data[data['Year'] <= 2023]
    test  = data[data['Year'] == 2024]

    if len(test) == 0:
        print(f"  No 2024 data found — skipping {name}")
        continue

    train_enc, encoders = encode(train[features + [TARGET]], fit=True)
    test_enc,  _        = encode(test[features + [TARGET]],  fit=False, encoders=encoders)

    X_train = train_enc[features]
    y_train = train_enc[TARGET]
    X_test  = test_enc[features]
    y_test  = test_enc[TARGET]

    model = GradientBoostingRegressor(
        n_estimators=300,
        max_depth=5,
        learning_rate=0.05,
        subsample=0.8,
        min_samples_leaf=10,
        random_state=42,
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # clip negatives — delay can't be < 0
    y_pred = np.clip(y_pred, 0, None)

    r2   = r2_score(y_test, y_pred)
    mae  = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    # cost estimation
    cost_per_min          = COST_PER_MIN[name]
    avg_predicted_cost    = y_pred.mean() * cost_per_min
    total_predicted_cost  = y_pred.sum()  * cost_per_min

    print(f"  Rows (train): {len(X_train):,}  |  Rows (test): {len(X_test):,}")
    print(f"  R²  : {r2:.4f}")
    print(f"  MAE : {mae:.2f} min")
    print(f"  RMSE: {rmse:.2f} min")
    print(f"\n  Cost per minute  : ${cost_per_min:.2f} CAD")
    print(f"  Avg cost/incident: ${avg_predicted_cost:.2f} CAD")
    print(f"  Total est. cost  : ${total_predicted_cost:,.0f} CAD  (test period)")

    # top 10 feature importances
    fi = pd.Series(model.feature_importances_, index=features)
    print("\n  Top 10 features:")
    top10 = fi.sort_values(ascending=False).head(10)
    for feat, imp in top10.items():
        print(f"    {feat:<22} {imp:.4f}")

    results[name] = {
        'model':                model,
        'encoders':             encoders,
        'features':             features,
        'r2':                   r2,
        'mae':                  mae,
        'rmse':                 rmse,
        'avg_cost_per_incident': avg_predicted_cost,
        'total_cost':           total_predicted_cost,
        'y_test':               y_test,
        'y_pred':               y_pred,
    }

# ─────────────────────────────────────────────────────────────────────────────
# 6. SUMMARY TABLE
# ─────────────────────────────────────────────────────────────────────────────
print("\n\n" + "=" * 54)
print("  SUMMARY")
print("=" * 54)

summary_rows = []
for name, v in results.items():
    summary_rows.append({
        'Mode':              name,
        'R²':                round(v['r2'],   4),
        'MAE (min)':         round(v['mae'],  2),
        'RMSE (min)':        round(v['rmse'], 2),
        'Avg cost/incident': f"${v['avg_cost_per_incident']:.2f}",
        'Total cost (test)': f"${v['total_cost']:,.0f}",
    })

summary = pd.DataFrame(summary_rows).set_index('Mode')
print(summary.to_string())

# ─────────────────────────────────────────────────────────────────────────────
# 7. PREDICT ON NEW DATA  (optional helper)
# ─────────────────────────────────────────────────────────────────────────────
def predict_delay_cost(mode, new_df):
    """
    Predict delay minutes and estimated cost for new input rows.

    Parameters
    ----------
    mode   : 'Bus', 'Subway', or 'Streetcar'
    new_df : DataFrame with the same feature columns used in training

    Returns
    -------
    DataFrame with added columns: predicted_delay_min, predicted_cost_cad
    """
    if mode not in results:
        raise ValueError(f"Mode '{mode}' not found in results.")

    r        = results[mode]
    features = r['features']
    encoders = r['encoders']
    model    = r['model']

    df = new_df.copy()
    df, _ = encode(df, encoders=encoders, fit=False)

    preds = np.clip(model.predict(df[features]), 0, None)
    df['predicted_delay_min']  = preds
    df['predicted_cost_cad']   = preds * COST_PER_MIN[mode]
    return df
```

```python
import requests
import pandas as pd
import io

station_id = 51459  # Toronto City Centre — from the URL in your screenshot
years = [2023, 2024, 2025]
all_weather = []

for year in years:
    for month in range(1, 13):
        url = (
            f"https://climate.weather.gc.ca/climate_data/bulk_data_e.html"
            f"?format=csv&stationID={station_id}"
            f"&Year={year}&Month={month}&Day=1&timeframe=1&submit=Download+Data"
        )
        r = requests.get(url)
        # skip header rows Environment Canada adds at the top
        try:
            df = pd.read_csv(io.StringIO(r.text), skiprows=0)
            all_weather.append(df)
            print(f"✓ {year}-{month:02d}")
        except Exception as e:
            print(f"✗ {year}-{month:02d} — {e}")

weather_raw = pd.concat(all_weather, ignore_index=True)
weather_raw.to_csv('weather_2023_2025.csv', index=False)
print("Done — saved to weather_2023_2025.csv")
print(weather_raw.columns.tolist())  # print columns so we can check names
```

```python
weather = weather_raw.rename(columns={
    'Year':                 'Year',
    'Month':                'Month',
    'Day':                  'Day',
    'Time (LST)':           'Hour',
    'Temp (°C)':            'Temp',
    'Precip. Amount (mm)':  'Precipitation',
    'Wind Spd (km/h)':      'WindSpeed',
    'Rel Hum (%)':          'Humidity',
    'Wind Chill':           'WindChill',
})

# keep only what we need
weather = weather[['Year', 'Month', 'Day', 'Hour', 
                   'Temp', 'Precipitation', 'WindSpeed', 
                   'Humidity', 'WindChill']]

# Hour is "00:00" string — extract integer
weather['Hour'] = pd.to_datetime(weather['Hour'], format='%H:%M').dt.hour

# fill nulls
weather['Precipitation'] = weather['Precipitation'].fillna(0)
weather['WindSpeed']      = weather['WindSpeed'].fillna(0)
weather['WindChill']      = weather['WindChill'].fillna(weather['Temp'])
weather['Humidity']       = weather['Humidity'].fillna(weather['Humidity'].median())
weather                   = weather.dropna(subset=['Temp'])  # temp is critical

# add snow flag — WindChill active + near-freezing = likely snow conditions
weather['IsSnowCondition'] = (
    (weather['Temp'] <= 2) & (weather['Precipitation'] > 0)
).astype(int)

print(weather.shape)
print(weather.isnull().sum())
weather.head()
```

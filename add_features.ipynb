```python
import pandas as pd
import numpy as np
df_bus = pd.read_csv('clean_bus.csv')
df_streetcar = pd.read_csv('clean_streetcar.csv')
df_sub = pd.read_csv('clean_sub.csv')
```

```python
allData = [df_bus, df_streetcar, df_sub]
```

```python
for a in allData:
    print(a.Hour.unique())
```

```python
#adding a flag if the hours were in rush hour
for a in allData:
    a['IsRushHour'] = np.where(
    a['Hour'].between(7, 9) | a['Hour'].between(16, 18),
    1, 0
)
```

```python
#adding a flag if the days were weekend
for a in allData:
    a['IsWeekend'] = np.where(
    a['DayOfWeek']>=5,
    1, 0
)
```

```python
#adding a flag if it was night hours
for a in allData:
    a['IsNight'] = np.where(
    a['Hour'].between(0, 5),
    1, 0
)
```

```python
choices = ['Winter', 'Spring', 'Summer', 'Fall']

for a in allData:
    conditions = [
        a['Month'].isin([12, 1, 2]),
        a['Month'].between(3, 5),
        a['Month'].between(6, 8),
        a['Month'].between(9, 11),
    ]
    a['Season'] = np.select(conditions, choices, default='Unknown')
```
